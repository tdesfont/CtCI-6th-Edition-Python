```python
import networkx as nx
import pdb
import random
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
def create_graph(E, N, undirected=True):
    """
    Function for graph creation
    :param E:
    :param V:
    :return:
    """
    graph = defaultdict(set)
    for i in range(E):
        n1 = random.randint(0, N)
        n2 = n1
        while n2 == n1:
            n2 = random.randint(0, N)
        graph[n1].add(n2)
        if undirected:
            graph[n2].add(n1)
    return graph
```

```python
graph = create_graph(E=30, N=20)
G = nx.Graph()
for source in graph:
    for target in graph[source]:
        G.add_edges_from([(source, target)])
nx.draw(G, with_labels=True)
```
